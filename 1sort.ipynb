```python
import pandas as pd

data = pd.read_csv('1.best-harga_bahan_pokok_jawa_timur_2023.csv', sep=';')
print(data)

def selection_sort(data, column):
    for i in range(len(data)):
        min_idx = i
        for j in range(i+1, len(data)):
            if data.loc[j, column] < data.loc[min_idx, column]:
                min_idx = j
        data.iloc[i], data.iloc[min_idx] = data.iloc[min_idx].copy(), data.iloc[i].copy()
    return data

selection_sort(data, 'harga_rata_rata')
print(data)
```

```python
import pandas as pd

data = pd.read_csv('1.best-harga_bahan_pokok_jawa_timur_2023.csv', sep=';')
print(data)

def counting_sort(data, column):
    max_val = int(data[column].max())
    count = [0] * (max_val + 1)
    output = [0] * len(data)

    for i in data[column]:
        count[int(i)] += 1

    for i in range(1, len(count)):
        count[i] += count[i - 1]

    for i in range(len(data) - 1, -1, -1):
        output[count[int(data.loc[i, column])] - 1] = data.loc[i]
        count[int(data.loc[i, column])] -= 1

    for i in range(len(data)):
        data.loc[i] = output[i]

    return data

counting_sort(data, 'harga_rata_rata')
print(data)
```
